```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression
%matplotlib inline

file_name="housing.csv"
df = pd.read_csv(file_name)
df.head()
```

```python
# Display the data types of each column
df.dtypes
```

```python
df.drop(["id", "Unnamed: 0"], axis=1, inplace=True)
df.describe()
```

```python
# Count the number of houses with unique floor values and convert to DataFrame
floor_counts = df['floors'].value_counts().to_frame()
floor_counts
```

```python
# Create a boxplot to compare price outliers for waterfront vs non-waterfront houses
sns.boxplot(x='waterfront', y='price', data=df)
plt.xlabel('Waterfront (0 = No, 1 = Yes)')
plt.ylabel('Price')
plt.title('Boxplot of House Prices by Waterfront View')
plt.show()
```

```python
# Use seaborn's regplot to visualize the correlation between sqft_above and price
sns.regplot(x='sqft_above', y='price', data=df)
plt.xlabel('Square Footage Above Ground')
plt.ylabel('Price')
plt.title('Regression Plot of Price vs. Sqft Above')
plt.show()
```

```python
# Fit a linear regression model to predict price using 'sqft_living'
X = df[['sqft_living']]
y = df['price']

lm = LinearRegression()
lm.fit(X, y)

# Calculate R^2
r2 = lm.score(X, y)
print(f"R^2: {r2}")
```

```python
# 选择特征和目标变量
features = [
    "floors", "waterfront", "lat", "bedrooms", "sqft_basement",
    "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"
]
X_multi = df[features]
y_multi = df["price"]

# 处理缺失值（用均值填充）
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_multi_imputed = imputer.fit_transform(X_multi)

# 拟合线性回归模型
lm_multi = LinearRegression()
lm_multi.fit(X_multi_imputed, y_multi)

# 计算R^2
r2_multi = lm_multi.score(X_multi_imputed, y_multi)
print(f"R^2: {r2_multi}")
```

```python
# 创建包含标准化、二次多项式特征和线性回归的pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

features = [
    "floors", "waterfront", "lat", "bedrooms", "sqft_basement",
    "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"
]
X_multi = df[features]
y_multi = df["price"]

# 处理缺失值
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_multi_imputed = imputer.fit_transform(X_multi)

# 创建pipeline
pipe = Pipeline([
    ('scale', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('linear', LinearRegression())
])

# 拟合模型
pipe.fit(X_multi_imputed, y_multi)

# 计算R^2
r2_poly = pipe.score(X_multi_imputed, y_multi)
print(f"R^2 (Polynomial Regression): {r2_poly}")
```

```python
# 导入 Ridge 回归
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split

# 特征和目标变量
features = [
    "floors", "waterfront", "lat", "bedrooms", "sqft_basement",
    "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"
]
X = df[features]
y = df["price"]

# 处理缺失值
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=1)

# 创建并拟合 Ridge 回归模型
ridge = Ridge(alpha=0.1)
ridge.fit(X_train, y_train)

# 计算测试集上的 R^2
r2_ridge = ridge.score(X_test, y_test)
print(f"R^2 (Ridge Regression, alpha=0.1): {r2_ridge}")
```

```python
# 对训练集和测试集做二次多项式变换，并用 Ridge 回归拟合
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer

features = [
    "floors", "waterfront", "lat", "bedrooms", "sqft_basement",
    "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"
]
X = df[features]
y = df["price"]

# 处理缺失值
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=1)

# 多项式特征变换（2阶）
poly = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Ridge 回归
ridge_poly = Ridge(alpha=0.1)
ridge_poly.fit(X_train_poly, y_train)

# 计算测试集上的 R^2
r2_poly_ridge = ridge_poly.score(X_test_poly, y_test)
print(f"R^2 (Ridge Regression with Polynomial Features, alpha=0.1): {r2_poly_ridge}")
```
