```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression
%matplotlib inline

file_name="housing.csv"
df = pd.read_csv(file_name)
df.head()
```

```python
# Display the data types of each column
df.dtypes
```

```python
df.drop(["id", "Unnamed: 0"], axis=1, inplace=True)
df.describe()
```

```python
# Count the number of houses with unique floor values and convert to DataFrame
floor_counts = df['floors'].value_counts().to_frame()
floor_counts
```

```python
# Create a boxplot to compare price outliers for waterfront vs non-waterfront houses
sns.boxplot(x='waterfront', y='price', data=df)
plt.xlabel('Waterfront (0 = No, 1 = Yes)')
plt.ylabel('Price')
plt.title('Boxplot of House Prices by Waterfront View')
plt.show()
```

```python
# Use seaborn's regplot to visualize the correlation between sqft_above and price
sns.regplot(x='sqft_above', y='price', data=df)
plt.xlabel('Square Footage Above Ground')
plt.ylabel('Price')
plt.title('Regression Plot of Price vs. Sqft Above')
plt.show()
```

```python
# Fit a linear regression model to predict price using 'sqft_living'
X = df[['sqft_living']]
y = df['price']

lm = LinearRegression()
lm.fit(X, y)

# Calculate R^2
r2 = lm.score(X, y)
print(f"R^2: {r2}")
```

```python
# é€‰æ‹©ç‰¹å¾å’Œç›®æ ‡å˜é‡
features = [
    "floors", "waterfront", "lat", "bedrooms", "sqft_basement",
    "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"
]
X_multi = df[features]
y_multi = df["price"]

# å¤„ç†ç¼ºå¤±å€¼ï¼ˆç”¨å‡å€¼å¡«å……ï¼‰
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_multi_imputed = imputer.fit_transform(X_multi)

# æ‹Ÿåˆçº¿æ€§å›žå½’æ¨¡åž‹
lm_multi = LinearRegression()
lm_multi.fit(X_multi_imputed, y_multi)

# è®¡ç®—R^2
r2_multi = lm_multi.score(X_multi_imputed, y_multi)
print(f"R^2: {r2_multi}")
```

```python
# åˆ›å»ºåŒ…å«æ ‡å‡†åŒ–ã€äºŒæ¬¡å¤šé¡¹å¼ç‰¹å¾å’Œçº¿æ€§å›žå½’çš„pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

features = [
    "floors", "waterfront", "lat", "bedrooms", "sqft_basement",
    "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"
]
X_multi = df[features]
y_multi = df["price"]

# å¤„ç†ç¼ºå¤±å€¼
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_multi_imputed = imputer.fit_transform(X_multi)

# åˆ›å»ºpipeline
pipe = Pipeline([
    ('scale', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('linear', LinearRegression())
])

# æ‹Ÿåˆæ¨¡åž‹
pipe.fit(X_multi_imputed, y_multi)

# è®¡ç®—R^2
r2_poly = pipe.score(X_multi_imputed, y_multi)
print(f"R^2 (Polynomial Regression): {r2_poly}")
```

```python
# å¯¼å…¥ Ridge å›žå½’
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split

# ç‰¹å¾å’Œç›®æ ‡å˜é‡
features = [
    "floors", "waterfront", "lat", "bedrooms", "sqft_basement",
    "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"
]
X = df[features]
y = df["price"]

# å¤„ç†ç¼ºå¤±å€¼
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# åˆ’åˆ†è®­ç»ƒé›†å’Œæµ‹è¯•é›†
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=1)

# åˆ›å»ºå¹¶æ‹Ÿåˆ Ridge å›žå½’æ¨¡åž‹
ridge = Ridge(alpha=0.1)
ridge.fit(X_train, y_train)

# è®¡ç®—æµ‹è¯•é›†ä¸Šçš„ R^2
r2_ridge = ridge.score(X_test, y_test)
print(f"R^2 (Ridge Regression, alpha=0.1): {r2_ridge}")
```

```python
# å¯¹è®­ç»ƒé›†å’Œæµ‹è¯•é›†åšäºŒæ¬¡å¤šé¡¹å¼å˜æ¢ï¼Œå¹¶ç”¨ Ridge å›žå½’æ‹Ÿåˆ
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer

features = [
    "floors", "waterfront", "lat", "bedrooms", "sqft_basement",
    "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"
]
X = df[features]
y = df["price"]

# å¤„ç†ç¼ºå¤±å€¼
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# åˆ’åˆ†è®­ç»ƒé›†å’Œæµ‹è¯•é›†
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=1)

# å¤šé¡¹å¼ç‰¹å¾å˜æ¢ï¼ˆ2é˜¶ï¼‰
poly = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Ridge å›žå½’
ridge_poly = Ridge(alpha=0.1)
ridge_poly.fit(X_train_poly, y_train)

# è®¡ç®—æµ‹è¯•é›†ä¸Šçš„ R^2
r2_poly_ridge = ridge_poly.score(X_test_poly, y_test)
print(f"R^2 (Ridge Regression with Polynomial Features, alpha=0.1): {r2_poly_ridge}")
```
